```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y=make_regression(n_samples =4, n_features =1, n_informative = 1 , n_targets =1 , noise= 80 , random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
#let's apply OLS
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
LinearRegression(copy_X=True , fit_intercept=True , n_jobs = None)
```

```python
reg.coef_   #m slop
```

```python
reg.intercept_   #b
```

```python
plt.scatter(X,y) 
plt.plot(X, reg.predict(X) , color = 'red')
```

```python
# Let's apply Gradient Descent assuming slope is constant = 78.35 
# and let's assume the starting value for intercept  b = 0 
y_pred = ((78.35*X)+0).reshape(4) #y = m*x+b 
```

```python
plt.scatter(X,y) 
plt.plot(X , reg.predict(X) , color = 'red' , label = 'OLS')
plt.plot(X,y_pred , color = "green" , label = 'b = 0')
plt.legend() 
plt.show() 
```

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])

# using ravel() function to flatten the arrayflattened_arr=arr.ravel()
flattened_arr = arr.ravel()

print(arr)
# Output: [[1 2 3]
#          [4 5 6]]

print(flattened_arr)
# Output: [1 2 3 4 5 6]
```

```python
m = 78.35 
b = 0 
loss_slope = -2*np.sum(y - m*X.ravel()-b)
loss_slope
```

```python
# Let's taKe learning rate  
lr = 0.1 
step_size = loss_slope*lr 
step_size
```

```python
# Calculating the new intercept 
b = b-step_size 
b
```

```python
y_pred1 = ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X , reg.predict(X) , color = 'black' , label = 'OLS')
plt.plot(X , y_pred1 , color = 'red' , label = 'b = {}'.format(b))
plt.plot(X,y_pred , color = 'yellow' , label = 'b = 0') 
plt.legend() 
plt.show() 
```

```python
#Iteration -2 

loss_slope = -2*np.sum(y - m*X.ravel() - b) 
loss_slope
```

```python
step_size = loss_slope*lr 
step_size
```

```python
b = b - step_size 
b 
```

```python
y_pred2 = ((78.35*X)+b).reshape(4) 

plt.scatter(X,y) 
plt.plot(X , reg.predict(X) , color = 'red' , label = 'OLS')
plt.plot(X , y_pred2 , color = 'green' , label = 'b = {}'.format(b)) 
plt.plot(X , y_pred1 , color = 'blue' , label = 'b = {}'.format(b))
plt.plot(X , y_pred , color = 'yellow' , label = 'b=  0')
plt.legend() 
plt.show() 
```

```python
#changing Dtype from object to datetime
df['Start Date']=pd.to_datetime(df['Start Date'])
```

```python
df.nunique()
```

```python
df.isnull().sum()
```

```python
#let's try to fill in the missing values of gender with the string 'no gender'
df['Gender'].fillna('No Gender',inplace=True)
```

```python
df.isnull().sum()
```

```python
#let's fill the senior management with the mode values
mode = df['Senior Management'].mode().values[0]
df['Senior Management']= df['Senior Management'].replace(np.nan, mode)
 
df.isnull().sum()
```

```python
#now,for the first name and team , we cananot fill the missing values arbitarty data, so let's drop all the rows contaning these missing values 

df=df.dropna(axis=0,how='any')
print(df.isnull().sum())
df.shape
```

```python
#now encoding the data

from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Gender']=lb.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
#dat visualization====> histogram==>salary

import seaborn as sns
import matplotlib.pyplot as plt
sns.histplot(x='Salary',data=df)
plt.show()
```

```python
#Boxplot==> salary nad team

import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x='Salary',y='Team',data=df)
plt.show()
```

```python
#sactter boxplot for data visualization

sns.scatterplot(x='Salary',y='Team',data=df,hue='Gender',size='Bonus %')
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.show()
```

```python
#pairplot
sns.pairplot(df,hue='Gender',height=2)
```

```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
row_data = requests.get("")   
soup = BeautifulSoup(row_data.content) 
```

```python
soup
```

```python
#extract name class and create df
x = soup.select("._4rR01T") 
name = [] 
for i in x : 
    name.append((i.contents[0]))
```

```python
x
```

```python
name
```

```python
df_Name = pd.DataFrame(name , columns = ['Details']) 
```

```python
df_Name
```

```python
#extract price 
x = soup.select("._30jeq3")
price = [] 
for i in x : 
    price.append((i.contents[0]))
```

```python
x
```

```python
price
```

```python
df_price = pd.DataFrame(price , columns = ['price'])
```

```python
df_price
```

```python
#extract reviews
x = soup.select("._3LWZlK") 
review = [] 
for i in x : 
    review.append((i.contents[0]))
```

```python
x
```

```python
review
```

```python
df_ratings = pd.DataFrame(review , columns = ['ratings'])
```

```python
df_ratings
```

```python
df = pd.concat((df_Name , df_price , df_ratings) , axis = 1)
```

```python
df=df.dropna()
```

```python
df
```

```python
#save the df to a csv format
df.to_csv('scarp_data.csv',index=False)
```

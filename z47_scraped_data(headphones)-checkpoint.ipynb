```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd
```

```python
row_data = requests.get("https://www.flipkart.com/search?q=headphone&sid=0pm%2Cfcn&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_5_na_na_na&as-pos=2&as-type=RECENT&suggestionId=headphone%7CHeadset&requestId=dccab3ac-27ed-4c5c-b5cb-dedf75ceacd8&as-searchtext=headphone&p%5B%5D=facets.rating%255B%255D%3D3%25E2%2598%2585%2B%2526%2Babove")
soup = BeautifulSoup(row_data.content) 
```

```python
soup
```

```python
#extract name class and create df
x = soup.select(".s1Q9rs") 
name = [] 
for i in x : 
    name.append((i.contents[0]))
```

```python
x
```

```python
name
```

```python
df_Name = pd.DataFrame(name , columns = ['Details']) 
df_Name
```

```python
#extract price 
x = soup.select("._30jeq3")
price = [] 
for i in x : 
    price.append((i.contents[0]))
```

```python
x
```

```python
price
```

```python
df_price = pd.DataFrame(price , columns = ['price'])
```

```python
df_price
```

```python
#extract reviews
x = soup.select("._3LWZlK") 
review = [] 
for i in x : 
    review.append((i.contents[0]))
```

```python
x
```

```python
review
```

```python
df_ratings = pd.DataFrame(review , columns = ['ratings'])
```

```python
df_ratings
```

```python
df = pd.concat((df_Name , df_price , df_ratings) , axis = 1)
```

```python
df=df.dropna()
```

```python
df
```

```python
#save the df to a csv format
df.to_csv('scarped_data.csv',index=False)
```
